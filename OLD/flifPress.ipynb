```python
if not isinstance(m, list):
    m=m.tolist()  ##convert to list
```

```python
## Need to pad the list with 0's so that all sublists are the same size
maxLength = max(len(x) for x in m) # find largest sublist in list
```

```python
#Now we iterate over each list in the series, if the length of each sublist matches the max length, we'll skip it, otherwise
#we will need to pad it with zeros
#This function works by padding the last sublist with 0's, lst being the immediate list and si being the max size

def padList(lst, si):
    if len(lst) < si:
        [lst.append(0) for i in range(si - len(lst))]
    return lst

[padList(x, maxLength) for x in m] #this pads each list horizontally with zeros
```
